```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data
def points_within_circle(radius, 
                         center=(0, 0),
                         number_of_points=100):
    center_x, center_y = center
    points = []
    for _ in range(number_of_points):
        r = radius * np.sqrt(np.random.random())
        theta = np.random.random() * 2 * np.pi
        x = center_x + r * np.cos(theta)
        y = center_y + r * np.sin(theta)
        points.append((x, y))
    return zip(*points)
```

```python
X = np.arange(0, 7)
fig, ax = plt.subplots()
oranges_x, oranges_y = points_within_circle(10, (5, 2), 300)
lemons_x, lemons_y = points_within_circle(1.9, (2, 5), 300)


ax.scatter(oranges_x, oranges_y, c="orange")
ax.scatter(lemons_x, lemons_y, c="y")

ax.plot(X, 4.5 / 5 * X, "g-", linewidth=3)
plt.show()
```
